```python
#solving a tridiagonal system of equations
import numpy as np
from scipy.linalg import solve_banded
n = 10
u = np.ones(n-1, dtype=float) 
d = np.ones(n, dtype=float) 
l = np.ones(n-1, dtype=float)

u = np.pad(u, (1,0), mode='constant', constant_values=0)
l = np.pad(l, (0,1), mode='constant', constant_values=0)
d = -d * 2.
ab = np.array([u, d, l])
b = np.random.rand(n)

x = solve_banded((1,1), ab, b)

A = np.zeros((n,n), dtype=float)

for i in range(n):
    A[i,i] = d[i]
    if i > 0:
        A[i-1,i] = u[i]
    if i < n-1:
        A[i+1,i] = l[i]
        
e = np.linalg.norm(A@x - b)
print(e)
```
