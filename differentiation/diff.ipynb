```python
#finite differentiation
def diff_forward(f, x, h):
    return (f(x+h) - f(x))/h
def diff_backward(f, x, h):
    return (f(x)-f(x-h))/h
def diff_center(f, x, h):
    return (f(x+h) - f(x-h)) / (2*h)
```

```python
import numpy as np

f = np.sin
h = 1e-10
diff_result = diff_forward(f, 0.0, h)

print(diff_result)
```

```python
df = lambda t: diff_forward(f, t, h)
d2f = lambda t: diff_backward(df, t, h)

print(df(0))
print(d2f(np.pi/2))
```

```python
from scipy.special import factorial
import numpy as np
def fdcoeff(k, xbar, x):
    n = len(x)
    A = np.ones((n, n))
    b = np.zeros(n)
    b[k] = 1
    for j in range(1,n):
        A[j,:] = (x - xbar)**j / factorial(j)
    c = np.linalg.solve(A,b)
    return c
```

```python
x = np.linspace(-0.1, 0.1, 5)
f = lambda x: x**3 + 5*x**2 + 3*x + 2
f_vec = np.vectorize(f)
fval =  f_vec(x)
coeff = fdcoeff(2, 0, x)
diff = np.dot(coeff, fval)
print(diff)

```
